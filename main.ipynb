```python
from Bio import SeqIO
from Bio.SeqUtils import GC
import matplotlib.pyplot as plt
```

```python
genes = SeqIO.parse('./genes.fasta', 'fasta')
```

```python
genes_list = [gene.seq for gene in genes]
```

```python
sizes = [len(gene) for gene in genes_list]
```

```python
plt.hist(sizes, bins= 15)
plt.title(f"{len(sizes)} genes sequences\nLengths {min(sizes)} to {max(sizes)}")
plt.xlabel('Sequence length (bp)')
plt.ylabel('Count')
plt.show()
```

```python
gc_values = sorted(GC(rec) for rec in genes_list)
```

```python
plt.plot(gc_values)
plt.title(f"{len(gc_values)} genees sequences\nGC%% {min(gc_values)} to {max(gc_values)}")
plt.xlabel("Genes")
plt.ylabel("GC%")
plt.show()
```

1) How many base pairs long are most of the sequences in genes.fasta

Most gene sequences are 3kbp - 5kbp long. 

2) What are the minimum and maximum GC% content

minimum: ~37%

maximum: ~66%
